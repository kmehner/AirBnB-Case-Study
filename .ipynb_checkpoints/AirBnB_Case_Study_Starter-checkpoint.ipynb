```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('./AB_NYC_2019.csv')
air_bnb
```

```python
air_bnb.dtypes
```

```python
air_bnb.isnull().sum()
```

```python
air_bnb.fillna({'reviews_per_month':0}, inplace=True)
air_bnb.reviews_per_month.isnull().sum()
```

```python
# How many neighbourhood groups / which shows up the most

# There are 5 neighbourhood groups
air_bnb.neighbourhood_group.unique()

# Manhattan shoows up the most
air_bnb.groupby('neighbourhood_group').count()['neighbourhood']
```

```python
# Are private rooms the most popular in Manhattan?

# The room types aree private, entire, and shared 

rooms_manhattan = air_bnb[['neighbourhood_group', 'room_type']]

top_rooms = rooms_manhattan[rooms_manhattan['neighbourhood_group'] == 'Manhattan']

# No, the most popular rooms are entire home/apt at 13,199.
# Private rooms are second most popular at 7,982
top_rooms.value_counts()
```

```python
# Who are the top 10 busiest hosts based on number of reviews per month?

# gathering the data we need 
hosts = air_bnb[['host_name','host_id','reviews_per_month']]

# top 10 busiest hosts
top_hosts = hosts.groupby(['host_name', 'host_id']).sum('reviews_per_month').reset_index()

top_hosts.sort_values('reviews_per_month', ascending = False).head(10)

```

```python
# Which neighborhood group has the highest avg price and plot the average prices in a bar chart

# Data we need
neighborhood_prices = air_bnb[['neighbourhood_group', 'price']]

avg_prices = neighborhood_prices.groupby('neighbourhood_group').mean().round(2).reset_index()

# Manhattan has the highest average prices
print(avg_prices)

avg_prices.plot.bar(x="neighbourhood_group", y='price', rot=40,  title="Avg Rental Price per Neighbourhood Group")

# get rid of output detailing graph and display
plt.show()
```

```python
# What is the most expensive neighbourhood in each neighbourhood_group

neighborhoods = air_bnb[['neighbourhood_group', 'price','neighbourhood']]

neighborhood = neighborhood_prices.groupby('neighbourhood_group').max()

print(neighborhood)

# The long way 
print(neighborhoods[neighborhoods['price']==2500]) # Bronx- Riverdale
print(neighborhoods[neighborhoods['price']==10000]) # Brooklyn - Greenpoint
                                                    # Manhattan - Upper West Side
                                                    # Queens - Astoria
print(neighborhoods[neighborhoods['price']==5000])  # Staten Island - Randall Manor
```

```python
# Part 2. # What is the most expensive neighbourhood in each neighbourhood_group

most_expensive = neighborhoods.groupby('neighbourhood_group').max()[['price', 'neighbourhood']]
# Correct prices but not printing the correct neighbourhood?
print(most_expensive)

# Testing against the results we found earlier 
# neighborhoods[neighborhoods['neighbourhood'] == 'Woodlawn'].max()
neighborhoods[neighborhoods['neighbourhood'] == 'Riverdale'].max()
```

```python
# Which host has the highest minimum rental? *minimal rental is based on price and minimum nights

air_bnb['minimum_rental'] = air_bnb['minimum_nights'] * air_bnb['price']

nights = air_bnb[['minimum_rental', 'host_id', 'host_name']]

night_max = nights['minimum_rental'].max()

nights[nights['minimum_rental'] == night_max]
```

```python
# Who has no (zero) availability with a review count of 100 or more

def availbility(row):
    if row['availability_365'] == 0 and row['number_of_reviews'] >= 100:
         return True
    else:
        return False

air_bnb['zero_availability'] = air_bnb.apply(availbility, axis=1)
air_bnb[air_bnb['zero_availability'] == True]
```

```python
# If each host rents all of their properties for the night, which 5 make the most that night?
# Assuming it is not based on availability

host_profit = air_bnb[['host_name', 'price']]
host_profit.groupby('host_name').sum('price').sort_values('price', ascending=False).head(5)
```

```python
# When did Danielle from Queens last receive a review?
dani = air_bnb[air_bnb['host_name'] == 'Danielle']
dani_queens = dani[dani['neighbourhood_group'] == 'Queens']

dani_queens.sort_values('last_review', ascending=False).head(1)[['host_name','last_review']]
```

```python
# bar chart showing the number of room types in the Bronx
bronx_rooms = air_bnb[air_bnb['neighbourhood_group'] == 'Bronx']
bronx_room_types = bronx_rooms.groupby('neighbourhood_group')['room_type'].value_counts()
print(bronx_room_types)

bronx_room_types.plot.bar(xlabel="Room Type", rot=10, ylabel="Listing type Frequency", title="Room Types in Bronx")
```
